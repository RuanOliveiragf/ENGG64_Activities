```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
img = cv.imread(r"flowers9.png") #leio a imagem

if img is None: #verifico se ela esta na pasta
    print("nao foi possivel ler a imagem")

plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()
```

```python
histograma_red = cv.calcHist(img, [2], None, [256], [0, 256])
histograma_green = cv.calcHist(img, [1], None, [256], [0, 256])
histograma_blue = cv.calcHist(img, [0], None, [256], [0,256])

hist = [histograma_blue, histograma_green, histograma_red]
colors = ['blue', 'green', 'red']

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))

ax1.plot(histograma_blue, color='blue')
ax1.set_title('Histograma - Azul')
ax1.set_xlim([0, 256])
ax1.grid(True, alpha=0.3)

ax2.plot(histograma_green, color='green')
ax2.set_title('Histograma - Verde')
ax2.set_xlim([0, 256])
ax2.grid(True, alpha=0.3)

ax3.plot(histograma_red, color='red')
ax3.set_title('Histograma - Vermelho')
ax3.set_xlim([0, 256])
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
img_bgr = cv.imread('flowers9.png')

if img_bgr is not None:
    img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)

    r, g, b = cv.split(img_rgb)
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))

    axs[0].imshow(r, cmap='gray')
    axs[0].set_title('Canal Vermelho (R)')
    axs[0].axis('off')

    axs[1].imshow(g, cmap='gray')
    axs[1].set_title('Canal Verde (G)')
    axs[1].axis('off')
    axs[2].imshow(b, cmap='gray')
    axs[2].set_title('Canal Azul (B)')
    axs[2].axis('off')

    plt.tight_layout()
    plt.show()
else:
    print("Erro ao carregar a imagem.")
```
