```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import sys

img = cv.imread("lena.pgm")

if img is None: #verifico se ela esta na pasta
    print("nao foi possivel ler a imagem")

plt.imshow(img)
plt.show()
```

# a) Brigthness

```python
img_double = img.astype(np.double)
img_double = img_double/255.
print(img_double.dtype)

def atalho(img):#pra usar no restante
  img_f = np.clip(img * 255, 0, 255).astype(np.uint8)
  return img_f

img_brightness = atalho(img_double + 1/4)
plt.imshow(img_brightness)
```

# b) Contrast

```python
img_contrast = atalho(img_double * 1.75)
plt.imshow(img_contrast)
```

# c) Negative

```python
img_negative = atalho(1 - img_double )
plt.imshow(img_negative)
```

# d) Posterization

```python
n = 6
img_posterization = atalho(np.floor(img_double * n)/n)
plt.imshow(img_posterization)
```
