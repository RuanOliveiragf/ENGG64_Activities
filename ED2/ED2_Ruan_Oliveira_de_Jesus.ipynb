```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import sys
%matplotlib inline

img = cv.imread(r"penguins.png") #leio a imagem

if img is None: #verifico se ela esta na pasta
    print("nao foi possivel ler a imagem")

plt.imshow(img, cmap = 'gray')
plt.axis('off')
plt.show() #ploto a imagem para conferir
```

# Questo 1) Leia a imagem e obtenha o histograma da imagem

```python
hist = cv.calcHist([img], [0], None, [256], [0, 256])

#plot do histograma
plt.figure(figsize=(10, 4))
plt.plot(hist)
plt.title("Histograma da Imagem")
plt.xlabel("Nvel de Cinza (0=Preto, 255=Branco)")
plt.ylabel("N de Pixels (Frequncia)")
plt.grid(True)
plt.show()
```

# Questo 2) Determine o nvel de cinza que ocorre com mais frequncia na imagem

```python
#np.argmax() retorna o NDICE o nvel de cinza que tem o maior valor
nivel_mais_frequente = np.argmax(hist)
print(f"O nvel de cinza que ocorre com mais frequncia : {nivel_mais_frequente}")
```

# Questo 3) Inspecione visualmente o histograma da imagem e determine um limite adequado para obter uma imagem binaria sem perda de informaes relevantes na imagem.

```python
#o histograma tem dois picos
#vou usar o ponto mais baixo como um limiar
limiar_escolhido = 135

#cv.threshold(imagem, limiar, valor_max, tipo)
#qualquer pixel > 135 se torna 255
# Qualquer pixel <= 135 se torna 0
ret, img_binaria = cv.threshold(img, limiar_escolhido, 255, cv.THRESH_BINARY)

print(f"Limiar (Threshold) escolhido visualmente: {limiar_escolhido}")
```

# Questo 4 e 5) Calcule a imagem binria lgica correspondente a esse limite utilizando uma funo de threshold. Plote a imagem binria lgica

```python
plt.imshow(img_binaria, cmap='gray')
plt.title(f"Imagem Binria (Limiar = {limiar_escolhido})")
plt.axis('off')
plt.show()
```
