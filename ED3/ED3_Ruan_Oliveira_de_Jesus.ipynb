```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import sys

img = cv.imread("penguins.png")


Iu = cv.Sobel(img, cv.CV_8U, 1, 0, ksize=3)
Iv = cv.Sobel(img, cv.CV_8U, 0, 1, ksize=3)


titles = ["sobrel vertical", "sobrel horizontal"]

plt.figure(figsize=(13, 13))
for i, grad_img in enumerate([Iv, Iu]):
    plt.subplot(1, 2, i+1)
    plt.title(titles[i])
    plt.imshow(cv.cvtColor(grad_img, cv.COLOR_BGR2RGB))

plt.tight_layout()
plt.show()

I = np.sqrt((Iv.astype(float)**2)+(Iu.astype(float)**2))
I = (I / I.max() * 255).astype(np.uint8)


plt.figure(figsize=(8, 8))
plt.title("Imagem das Bordas (Magnitude)")
plt.imshow(cv.cvtColor(I, cv.COLOR_BGR2RGB))
plt.show()
```
